```python
from gensim.models import Word2Vec, KeyedVectors
model = KeyedVectors.load_word2vec_format('../../../GoogleNews-vectors-negative300.bin.gz', binary=True, limit=200000)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from nltk import word_tokenize, pos_tag,RegexpTokenizer
import numpy as np

sent_pad_len = 50
```

```python
def preprocessing_convert2BOW(sentence):
    
    tokenizer = RegexpTokenizer(r'\w+')
#     Remove common words and change to lower
    common_words = set('for a of the and to in ,'.split()) 
    if sentence is not None:
        words = tokenizer.tokenize(sentence);
        bow = [w.lower() for w in words]
        return bow
    
    return []

def get_enhanced_wordvec(w, sent_pair):
    
    vectors = np.zeros(300)
    if w in model.vocab:
        vectors = model[w]
        
#   words in s1 overlap with s2
#     if w.isalpha() and w in sent_pair:
#         np.append(vectors,[1])
#     else:
#         np.append(vectors,[0])
#     if not w.isalpha() and w in sent_pair:
#         np.append(vectors,[1])
#     else:
#         np.append(vectors,[0])
        
    
    return vectors

def add_one_hot_vector_pos(sentence):
    pass
    

def convert2BOWV(s1, s2):
    s1_word_list = preprocessing_convert2BOW(s1)
    s2_word_list = preprocessing_convert2BOW(s2)
    s1_Zlen = sent_pad_len - len(s1_word_list)
    s2_Zlen = sent_pad_len - len(s2_word_list)
    s1_wv_list = []
    s2_wv_list = []
    
    for w in s1_word_list:
        s1_wv_list.append(get_enhanced_wordvec(w, s2)) 
    if s1_Zlen > 0:
        for i in range(s1_Zlen):
            s1_wv_list.append(np.zeros(300))
    
    for w in s2_word_list:
        s2_wv_list.append(get_enhanced_wordvec(w, s1)) 
        
    if s2_Zlen > 0:
        for i in range(s2_Zlen):
            s2_wv_list.append(np.zeros(300))
          
        
    return (s1_wv_list,s2_wv_list)

# sum of all word vectors
def getSentenceEmbedings(s1, s2):
    
    (s1_wv_list, s2_wv_list) = convert2BOWV(s1,s2)
    sv1 = s1_wv_list[0]
    sv2 = s2_wv_list[0]
    
    for i in range(len(s1_wv_list)):
        if i != 0:
            sv1 = np.add(sv1,s1_wv_list[i])
    
    for i in range(len(s2_wv_list)):
        if i != 0:
            sv2 = np.add(sv2,s2_wv_list[i])
            
    return sv1, sv2               

sent1 = "Nobody is riding the bicycle on one wheel"
sent2 = "A person in a black jacket is doing tricks on a motorbike"
sv1,sv2 = getSentenceEmbedings(sent1,sent2)

sv1 = sv1.reshape(1, -1)
sv2 = sv2.reshape(1, -1)
```
